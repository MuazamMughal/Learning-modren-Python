## file handling


```python
from typing import TextIO
# establiing the conation to the file using open
data: TextIO  = open("./abc.txt")
print(data.read())
# now closing the conection
data.close()
```

```python
#with  is for autioatialy close the file after working in it
with open("./abc.txt") as file:
    print(type(file))
    print(file.read()) #reading the whole lines
   
    
```

```python
with open("./abc.txt") as file: #type = TextIO
    print(type(file))
  
    # now reading desire no of lines
    #print(file.readlines()[:3])
    # now reading the characters
    print(file.readline()[:24]) # reading the first line 
     
    print(file.readline()[:24]) #reading the second


    #the readlines reads lines and the readline is reading the character
  
    
    
```

```python
with open("./abc.txt") as file:
    print(type(file))
    print(file.readline() ,end="") #end is for read a line and then end 
    print(file.readline() ,end="")
    print(file.readline() ,end="")
```
