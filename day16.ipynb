```python
#selectin with boolean
import numpy as np
from nptyping import NDArray,Shape,Bool 
#this is an our atuaal from which we select
select :NDArray[Shape["20"],Any]= np.array([1,2,3,4])
boolData : NDArray[Shape["10"] , Bool] = np.array([True,False,True,False])
#now passing the bool to select
trueResult=select[boolData]
print(f"thes are the true owns{trueResult}")

# now more maupulatin f the narray
modulas = select[select % 2 == 0]
print(f"these are the values where 2 modulas is zero{modulas}")
```

```python
#important functions in np to remaember
#finding the required desire number in the long array
bankData : NDArray[Shape["1"], Any] = np.random.randint(1,100 ,20)
# now we gona find  the desire number is present in data

findings= np.in1d(select , bankData )
# in onedementional data (to find in , this data) will gieve bool
print(f"we find in the large random data array{findings}")
```

```python
display(bankData)
display(select)
print(f" finding intersected data{np.intersect1d(select ,bankData )}")
```

```python
#where use as findng between 2 array 
x:NDArray[Shape["10"], Any] = np.array([1,2,3,7,8])
y : NDArray[Shape["10"],Any] = np.array([1,6,8,9,4])
display(x)
display(y)
# NOw we gona compair the both array using the array
compair = np.where(x>y , x,y)
display(compair)


            
```

## Create np array

```python
#2d array
a : NDArray[Shape["*,*"] , Any] = np.array([[1,2,3,3],
                                                     [12,3,4,5]])
display(a)

b : NDArray[Shape["*,*"] , Any] = np.array([[1,2],
                                                     [12,4]])
display(b)
c : NDArray[Shape["*,*"] , Any] = np.array([["s " ,"q"],
                                                     [12,4]])
display(c)
```

 ##  creating multidimention arary

```python
d : NDArray[Shape["Size"], Any] = np.arange(1,3)
display(d)
e : NDArray[Shape["Size"], Any] = np.arange(1,9)
display(e)
#Now 2 d array

f : NDArray[Shape["Size,Size"], Any]= np.arange(3*3).reshape(3,3)
display(f)
g : NDArray[Shape["Size,Size"], Any]= np.arange(5*3).reshape(5,3)
display(g)
```

```python
#making larger arrays
h : NDArray[Shape["Size,Size ,Size"], Any]= np.arange(5*1*3).reshape(5,1,3)
display(h)
#5 is the nnober of total matriz or set of array
#1 is rows
#3 is number of colomns
```
