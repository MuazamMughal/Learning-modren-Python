```python
## now applay coustom function o multiplayl  coloums
def fun_obtained(s1:int,s2:int,s3:int,s4:int,s5:int)->int:
    return s1+s2+s3+s4+s5


#using lambda function
df1["obtained"] = df1[['s1','s2','s3','s4','s5']].apply( lambda x:fun_obtained(*x) ,axis=1)
display(df1)
```

```python
#now multi input with multi output senario
def resultes(s1:int,s2:int,s3:int,s4:int,s5:int)->tuple[int,int,float,str]:
    total : int =500
    obtained :int = s1+s2+s3+s4+s5
    per : float = obtained / total *100
    Grades:str = ''
    if per >=90:
         Grades ="A+"
    elif per >=80:
        Grades = "A"
    elif per >=70:
        Grades = "B"
    elif per >=60:
        Grades = "C"
    elif per >=50:
        Grades = "D"
    else:
        Grades = "Fail"
    
    return total,obtained,per,Grades

df1[["total","obtained","per","Grades"]] = df1[['s1','s2','s3','s4','s5']].apply(lambda x:resultes(*x),
                                                                               axis=1,
                                                                                result_type='expand' )
display(df1)
```
