```python
# now learning the data frames 
# dataframe whan we conect 1 or more than 1 data sets togaether

s1 : pd.Series = pd.Series([1,2,3,4,5] , name ="student_id")
s2 : pd.Series = pd.Series([10,20,30,40,50] , name ="markes")
s3 : pd.Series = pd.Series(['ali','ahmed','uzi','ak','m24'],name ="student_id")

# now combining them to make data frame
df: pd.DataFrame = pd.DataFrame({'student_id':s1 , 
                                 "score":s2,
                                 "student_name":s3})
display(df)

#another way of combining is concatinate
df2 :pd.DataFrame = pd.concat([s1,s2,s3] ,axis=1)
display(df2)
```

```python
import pandas as pd
import numpy as np
from nptyping import Shape,NDArray
from typing import Any
s2: pd.Series=pd.Series([1,2,3,4,5])
display(s2)
display(f'this is the basic slicing  {s2[1]}')
display(f'this is the basic slicing ')
display(s2[1:4])

s1: pd.Series = pd.Series([1,2,3,4,5] ,index=['a','b','c','d','e'])
display(s1)
```

```python
#using the iloc (index location)
display(s1)
# now we gona applay silicing using the iloc working on num index default
display(s1.iloc[1:4])
#now using the loc wich using the atual index
display(s1.loc['a':'c'])
#now slicing using the iat  only one cell value and update using the default numeric index 
display(s1.iat[2])
# slicing using the at using the atual index get value then also update
display(s1.at["d"])
```
